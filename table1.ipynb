```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency
```

```python
def get_pretty_pecent(a, b):
    #return round(100 * float(a)/float(b), 2)
    return "{:.2%}".format(a/b)
def get_chi(a, b):
    ctab = pd.crosstab(a, b)
    #print(ctab)
    chi2, p, dof, ex = chi2_contingency(ctab)
    if (p < 0.0005):
        p = '<0.0005';
    return p
```
