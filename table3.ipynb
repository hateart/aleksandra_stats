```python
def get_pretty_pecent(a, b):
    if b > 0:
        return round(100 * float(a)/float(b), 2)
    return 0
    #return "{:.2%}".format(a/b)
def get_chi(a, b):
    ctab = pd.crosstab(a, b)
    #print(ctab)
    chi2, p, dof, ex = chi2_contingency(ctab)
    if (p < 0.0005):
        p = '<0.0005';
    return p
```
