```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("./prepared_data.csv", index_col=False)
data = data.drop(data[data['StageGroup'].isnull()].index)
#data = data.dropna(axis=0)
len(data)
```

```python
data.head()
```

```python
#a = data.drop(data[data['Education'].isnull()].index)
a = data
a['Education'].fillna("EducationUnknown", inplace = True)
a = a.pivot(index = None, columns=['Education'], values=['Education'])
a = a.fillna(0)
a = a.replace(['Secondary', 'Secondary_vocational', 'University&higher', 'Basic&Primary', 'EducationUnknown'], 1)
a.head()
```

```python
#b = data.drop(data[data['Marital Status'].isnull()].index)
b = data
b['Marital Status'].fillna("MaritalStatusUnknown", inplace = True)
b = b.pivot(index = None, columns=['Marital Status'], values=['Marital Status'])
b = b.fillna(0)
b = b.replace(['widow', 'married', 'divorced', 'single', 'MaritalStatusUnknown'], 1)
b.head()
```

```python
c = data
c['AgeGroup'].fillna("AgeGroupUnknown", inplace = True)
c = data.pivot(index = None, columns=['AgeGroup'], values=['AgeGroup'])
c = c.fillna(0)
c = c.replace(['<= 29', '30-44', '45-59', '60-74', '>= 75', 'AgeGroupUnknown'], 1)
c.head()
```

```python
#d = data.drop(data[data['Nationality'].isnull()].index)
d = data
d['Nationality'].fillna("NationalityUnknown", inplace = True)
d = d.pivot(index = None, columns=['Nationality'], values=['Nationality'])
d = d.fillna(0)
d = d.replace(['Estonian', 'Russian', 'Other', 'NationalityUnknown'], 1)
d.head()
```

```python
#e = data.drop(data[data['Residence'].isnull()].index)
e = data
e['Residence'].fillna("ResidenceUnknown", inplace = True)
e = e.pivot(index = None, columns=['Residence'], values=['Residence'])
e = e.fillna(0)
e = e.replace(['South', 'North-East', 'Central', 'West', 'North', 'ResidenceUnknown'], 1)
e.head()
```

```python
#f = data.drop(data[data['CodeGroup'].isnull()].index)
f = data
f['CodeGroup'].fillna("CodeGroupUnknown", inplace = True)
f = f.pivot(index = None, columns=['CodeGroup'], values=['CodeGroup'])
f = f.rename(columns={"Other": "CodeOther"})
f = f.fillna(0)
f = f.replace(['SCC', 'AC', 'Other', 'CodeGroup'], 1)
f.head()
```

```python
final = a
final = final.join(b) 
final = final.join(c)
final = final.join(d) 
final = final.join(e) 
final = final.join(f) 
final.columns = final.columns.droplevel()
final['StageGroup'] = data['StageGroup'].astype(int)
```

```python
final.head()
```

```python
final.to_csv('data_for_poisson.csv', index=False)
```
